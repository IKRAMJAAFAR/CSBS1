# Dataset Reading & Extracting

```python
import pandas as pd
```

```python
dataset = pd.read_csv('Datasets\dataset.csv',index_col=0)
dataset
```

```python
dataset['timestamp'] = pd.to_datetime(dataset['timestamp'])
dataset.reset_index(inplace=True)
dataset.set_index('timestamp', inplace=True)

dataset.head()
```

```python
dataset['dayofweek'] = dataset.index.day_of_week
dataset['hour'] = dataset.index.hour + dataset.index.minute / 60

dataset
```

# Uploading Extracted Dataset

```python
dataset.to_csv("Datasets\dataset_extracted.csv")

check = pd.read_csv('Datasets\dataset_extracted.csv')

check
```
